```python
import numpy as np
import tensorflow as tf
import cv2
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import load_model
from skimage.transform import resize
import matplotlib.pyplot as plt


counter = 0
lastPressedKey = ''

#dodavanje kamere
camera = cv2.VideoCapture(0)
frames = []

while(1):
    #čitanje videa po frameovima
    _, imageFrame = camera.read()
    #pretvaranje slike iz BGR(RGB color space) u HSV(hue-saturation-value)
    hsvFrame = cv2.cvtColor(imageFrame, cv2.COLOR_BGR2HSV)

    #POSTAVLJANJE RANGE-ova BOJA I DEFINIRANJE MASKI
    white_lower = np.array([94, 80, 2], np.uint8)
    white_upper = np.array([120, 255, 255], np.uint8)
    white_mask = cv2.inRange(hsvFrame, white_lower, white_upper)

    kernal = np.ones((5, 5), "uint8")

    #Blue color
    blue_mask = cv2.dilate(white_mask, kernal)
    res_blue = cv2.bitwise_and(imageFrame, imageFrame,
                            mask = blue_mask)

    #Stvaranje konture za praćenje plave boje
    contours, hierarchy = cv2.findContours(blue_mask,
                                        cv2.RETR_TREE,
                                        cv2.CHAIN_APPROX_SIMPLE)

    boxes = []

    for pic, contour in enumerate(contours):
        area = cv2.contourArea(contour)
        if(area > 300):
            x, y, w, h = cv2.boundingRect(contour)
            boxes.append((x, y, w, h))
            xmin = x
            xmax = xmin + w
            ymin = y
            ymax = ymin + h
            #Crtanje okvira
            imageFrame = cv2.rectangle(imageFrame, (x, y),
                                (x + w, y + h),
                                (255, 0, 0), 2)
            cv2.putText(imageFrame, 'picture', (x, y),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.0,
                        (255, 0, 0))

    cv2.imshow("Color Detection in Real-Time", imageFrame)
    cv2.imshow("BlackNWhite", white_mask)

    #RIGHT HAND
    if cv2.waitKey(1) & 0xFF == ord('0'):
        frames.append((white_mask, boxes))
        if lastPressedKey != '0':
            lastPressedKey = '0'
            counter = getLastIndex('zero')
        saveImage('zero',counter)
        print(counter)
    elif cv2.waitKey(1) & 0xFF == ord('1'):
        frames.append((white_mask, boxes))
        if lastPressedKey != '1':
            lastPressedKey = '1'
            counter = getLastIndex('one')
        saveImage('one',counter)
    elif cv2.waitKey(1) & 0xFF == ord('2'):
        frames.append((white_mask, boxes))
        if lastPressedKey != '2':
            lastPressedKey = '2'
            counter = getLastIndex('two')
        saveImage('two',counter)
    elif cv2.waitKey(10) & 0xFF == ord('3'):
        frames.append((white_mask, boxes))
        if lastPressedKey != '3':
            lastPressedKey = '3'
            counter = getLastIndex('three')
        saveImage('three',counter)
    elif cv2.waitKey(100) & 0xFF == ord('4'):
        frames.append((white_mask, boxes))
        if lastPressedKey != '4':
            lastPressedKey = '4'
            counter = getLastIndex('four')
        saveImage('four',counter)
    elif cv2.waitKey(1) & 0xFF == ord('5'):
        frames.append((white_mask, boxes))
        if lastPressedKey != '5':
            lastPressedKey = '5'
            counter = getLastIndex('five')
        saveImage('five',counter)
#LEFT HAND
    if cv2.waitKey(1) & 0xFF == ord('6'):
        frames.append((white_mask, boxes))
        if lastPressedKey != '6':
            lastPressedKey = '6'
            counter = getLastIndex('six')
        saveImage('six',counter)
        print(counter)
    elif cv2.waitKey(1) & 0xFF == ord('7'):
        frames.append((white_mask, boxes))
        if lastPressedKey != '7':
            lastPressedKey = '7'
            counter = getLastIndex('seven')
        saveImage('seven',counter)
    elif cv2.waitKey(1) & 0xFF == ord('8'):
        frames.append((white_mask, boxes))
        if lastPressedKey != '8':
            lastPressedKey = '8'
            counter = getLastIndex('eight')
        saveImage('eight',counter)
    elif cv2.waitKey(1) & 0xFF == ord('9'):
        frames.append((white_mask, boxes))
        if lastPressedKey != '9':
            lastPressedKey = '9'
            counter = getLastIndex('nine')
        saveImage('nine',counter)
    elif cv2.waitKey(1) & 0xFF == ord('t'):
        frames.append((white_mask, boxes))
        if lastPressedKey != 't':
            lastPressedKey = 't'
            counter = getLastIndex('ten')
        saveImage('ten',counter)
    elif cv2.waitKey(1) & 0xFF == ord('e'):
        frames.append((white_mask, boxes))
        if lastPressedKey != 'e':
            lastPressedKey = 'e'
            counter = getLastIndex('eleven')
        saveImage('eleven',counter)
    elif cv2.waitKey(1) & 0xFF == ord('b'):
        frames.append((white_mask, boxes))
        if lastPressedKey != 'b':
            lastPressedKey = 'b'
            counter = getLastIndex('background')
        saveImage('background',counter)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        camera.release()
        cv2.destroyAllWindows()
        break




```
